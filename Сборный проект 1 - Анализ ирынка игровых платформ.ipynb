## Описание проекта

Для анализа предоставлены данные  интернет-магазина «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Необходимо выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании. Для анализа предоставленны данные c 1980 года до 2016 года. 
В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

**Необходимо:**
- Провести предподготовку данных и посчитать суммарные продажи игр во всех регионах.
- Провести исследовательский анализ данных
- Проанализировать, сколько игр выпускалось в разные годы. Важны ли данные за все периоды?
- Проанализировать, как менялись продажи по платформам. -
- Проанализировать, как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков. 
- Проанализировать  общее распределение игр по жанрам. Что можно сказать о самых прибыльных жанрах? Выделяются ли жанры с высокими и низкими продажами?

**Необходимо составить портрет пользователя каждого региона**

Определить для пользователя каждого региона (NA, EU, JP):
- Самые популярные платформы (топ-5). Опишите различия в долях продаж.
- Самые популярные жанры (топ-5). 
- Влияет ли рейтинг ESRB на продажи в отдельном регионе?

**Проверить гипотезы:** 

- Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
- Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.


**Описание данных:**
- Name — название игры
- Platform — платформа
- Year_of_Release — год выпуска
- Genre — жанр игры
- NA_sales — продажи в Северной Америке (миллионы проданных копий)
- EU_sales — продажи в Европе (миллионы проданных копий)
- JP_sales — продажи в Японии (миллионы проданных копий)
- Other_sales — продажи в других странах (миллионы проданных копий)
- Critic_Score — оценка критиков (максимум 100)
- User_Score — оценка пользователей (максимум 10)
- Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

## Проверка гипотез:

#### Проверка гипорезы о равенстве средних пользовательских рейтингов на Xbox One и PC

Формулируем гипотезу:

- Н0 - Средние значения пользовательских рейтингов платформ Xbox и PC равны.
- Н1 - Средние значения пользовательских рейтингов платформ Xbox и PC  не равны.

```python
data_xbox_scor = data_current_period.query('platform == "XOne"')
```

```python
data_pc_scor = data_current_period.query('platform == "PC"')
```

```python
alpha = 0.05
```

```python
result = st.ttest_ind(data_xbox_scor['user_score'], data_pc_scor['user_score'], nan_policy='omit')
```

```python
f'p-value: {result.pvalue}'
```

```python
if result.pvalue < alpha:
    print('Отвергаем нулевую гипотезу H0  в пользу альтернативной гипотизы Н1')
else:
    print('Нет оснований отвергнуть нулевую гипотезу')
```


Данные не дают достаточное основание отвергнуть нулевую гипотезу для выбранного уровня значимости.
После проведения ttesta нулевая гипотезе говорящая о равенстве средних значений оценок пользователей Xbox и PC имеет высокую  вероятность.

Гипотеза о равенстве средних значений пользовательских рейтингов платформ Xbox и PC **не опровергнута.**
